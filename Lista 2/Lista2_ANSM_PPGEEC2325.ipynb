Considere uma função densidade de probabilidade (fdp) Normal bidimensional fX,Y(X,Y), com médias mx = 5, my= 3 e matriz de covariância P = [4 1.8;1.8 1]. Gere 100.000 amostras para essa distribuição. Então:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D   # necessário para 3D
from matplotlib import cm
```

```python
# parâmetros
mu = np.array([5.0, 3.0])       # médias
P = np.array([[4.0, 1.8],
              [1.8, 1.0]])      # matriz dada no enunciado
N = 100000

# gerar amostras
np.random.seed(0)
samples = np.random.multivariate_normal(mu, P, size=N)

# função densidade multivariada (2D)
def mvn_pdf(xy, mu, Sigma):
    # xy : array of shape (...,2); retorna pdf por ponto
    d = 2
    det = np.linalg.det(Sigma)
    inv = np.linalg.inv(Sigma)
    norm_const = 1.0 / (2*np.pi * np.sqrt(det))
    dif = xy - mu
    # se xy é grid shape (M,N,2)
    # calcula expoente: -0.5 * sum(dif @ inv * dif)
    # remodela apropriadamente
    # achata o cálculo e depois remodela novamente, se necessário
    flat = dif.reshape(-1, 2)
    expo = -0.5 * np.sum((flat @ inv) * flat, axis=1)
    pdf_vals = norm_const * np.exp(expo)
    return pdf_vals.reshape(dif.shape[0], -1) if False else pdf_vals.reshape(*dif.shape[:-1])

# preparar grid para plots
x_min, x_max = samples[:,0].min() - 1, samples[:,0].max() + 1
y_min, y_max = samples[:,1].min() - 1, samples[:,1].max() + 1
nx, ny = 120, 120
x = np.linspace(x_min, x_max, nx)
y = np.linspace(y_min, y_max, ny)
X, Y = np.meshgrid(x, y)
XY = np.stack([X, Y], axis=-1)   # shape (ny, nx, 2)

# calcular pdf no grid
# adaptando mvn_pdf para grid:
def mvn_pdf_grid(XY, mu, Sigma):
    ny, nx, _ = XY.shape
    flat = XY.reshape(-1, 2)
    det = np.linalg.det(Sigma)
    inv = np.linalg.inv(Sigma)
    norm_const = 1.0 / (2*np.pi * np.sqrt(det))
    expo = -0.5 * np.sum((flat @ inv) * flat, axis=1)
    pdf_vals = norm_const * np.exp(expo)
    return pdf_vals.reshape(ny, nx)

Z = mvn_pdf_grid(XY, mu, P)
```

a. Plote o Gráfico 3D de fX,Y(X,Y).
O gráfico 3D é obtido pela fórmula da densidade normal bivariada. O código abaixo produz esse gráfico.

```python
# a) Gráfico 3D
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=3, cstride=3, linewidth=0, antialiased=True, cmap=cm.viridis)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f_{X,Y}(x,y)')
ax.set_title('3D - densidade normal bivariada (N=100000 amostras)')
fig.colorbar(surf, shrink=0.6)
plt.show()
```

```python
# c) Comparação com o item 1.j
# matriz covariância resultante do caso j (theta = pi/4, V = [2X, Y] com X,Y~N(0,1))
theta_j = np.pi/4
Rj = np.array([[np.cos(theta_j), -np.sin(theta_j)],
               [np.sin(theta_j),  np.cos(theta_j)]])
CovV = np.array([[4.0, 0.0],
                 [0.0, 1.0]])
P_j = Rj @ CovV @ Rj.T
print("Covariância prevista para 1.j (theta=pi/4):\n", P_j)
print("Matriz P do enunciado:\n", P)
```
