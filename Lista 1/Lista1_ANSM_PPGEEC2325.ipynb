Questão 1

```python
import numpy as np
import matplotlib.pyplot as plt

num_amostras = 10000

for n in range(1, 13):
    # Gera n sequências de 10000 amostras de U(0,1)
    amostras_uniformes = np.random.uniform(0, 1, size=(num_amostras, n))

    # Soma as amostras para obter a distribuição de X
    X = np.sum(amostras_uniformes, axis=1)

    # Calcula e imprime os resultados teóricos e simulados
    media_simulada = np.mean(X)
    variancia_simulada = np.var(X)
    segundo_momento_simulado = np.mean(X**2)

    media_teorica = 0.5 * n
    variancia_teorica = n / 12
    segundo_momento_teorico = variancia_teorica + media_teorica**2

    print(f"--- n = {n} ---")
    print(f"Média: Teórica = {media_teorica:.4f}, Simulada = {media_simulada:.4f}")
    print(f"Variância: Teórica = {variancia_teorica:.4f}, Simulada = {variancia_simulada:.4f}")
    print(f"Segundo Momento: Teórico = {segundo_momento_teorico:.4f}, Simulada = {segundo_momento_simulado:.4f}")

    # Gera o gráfico da PDF (histograma)
    plt.figure(figsize=(10, 5))
    plt.hist(X, bins=50, density=True, label=f'PDF (n={n})', alpha=0.7)
    plt.title(f'Função Densidade de Probabilidade (PDF) para n={n}')
    plt.xlabel('X')
    plt.ylabel('Densidade')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Gera o gráfico da CDF (função de distribuição acumulada)
    plt.figure(figsize=(10, 5))
    sorted_X = np.sort(X)
    cdf = np.arange(1, num_amostras + 1) / num_amostras
    plt.plot(sorted_X, cdf, label=f'CDF (n={n})')
    plt.title(f'Função de Distribuição de Probabilidade (CDF) para n={n}')
    plt.xlabel('X')
    plt.ylabel('Probabilidade Acumulada')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define o número de amostras e o parâmetro lambda (taxa) da exponencial.
num_amostras = 10000
lambda_param = 1.0  # O valor de lambda pode ser alterado para teste

# --- Geração da Distribuição Exponencial a partir da Uniforme ---

# 1. Gerar amostras de uma distribuição uniforme U[0,1]
# O np.random.uniform() gera números aleatórios de forma uniforme.
amostras_uniformes = np.random.uniform(0, 1, num_amostras)

# 2. Aplicar a transformação inversa
# A fórmula é x = - (1/lambda) * ln(1 - u)
# np.log() calcula o logaritmo natural (ln).
amostras_exponenciais = (-1/lambda_param) * np.log(1 - amostras_uniformes)

# --- Análise e Visualização ---

# 1. Obter o histograma para aproximar a PDF
plt.figure(figsize=(10, 6))
plt.hist(amostras_exponenciais, bins=50, density=True, alpha=0.7,
         label='Histograma (Simulado)')

# 2. Plotar a PDF teórica para comparação
# A curva teórica é f(x) = lambda * e^(-lambda * x)
x_vals = np.linspace(0, max(amostras_exponenciais), 500)
pdf_teorica = lambda_param * np.exp(-lambda_param * x_vals)
plt.plot(x_vals, pdf_teorica, 'r-', linewidth=2,
         label='PDF Teórica')

plt.title(f'Distribuição Exponencial (lambda = {lambda_param})')
plt.xlabel('x')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()

# 3. Validar com as estatísticas
# A média teórica de uma exponencial é 1/lambda
media_teorica = 1 / lambda_param
media_simulada = np.mean(amostras_exponenciais)

# A variância teórica de uma exponencial é 1/(lambda^2)
variancia_teorica = 1 / (lambda_param**2)
variancia_simulada = np.var(amostras_exponenciais)

print(f"Média: Teórica = {media_teorica:.4f}, Simulada = {media_simulada:.4f}")
print(f"Variância: Teórica = {variancia_teorica:.4f}, Simulada = {variancia_simulada:.4f}")
```

Questão 4

a. Histogramas para 10, 100, 1.000 e 10.000 pontos.


b. Média e variância para os mesmos tamanhos de amostra.

A análise deve focar em como a média e a variância se comportam à medida que o número de pontos aumenta.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir os parâmetros e o número total de amostras
num_amostras_total = 10000
amostras_a_analisar = [10, 100, 1000, 10000] # Pontos para análise [cite: 22, 24]

# Gerar a sequência de 10.000 pontos da distribuição Normal N(0,1)
# np.random.normal(media, desvio_padrao, tamanho)
amostras_normais = np.random.normal(loc=0, scale=1, size=num_amostras_total)

# --- Seção a: Gerar e Analisar os Histogramas ---
print("--- Análise dos Histogramas ---")
for n in amostras_a_analisar:
    # Seleciona o subconjunto de amostras
    sub_amostras = amostras_normais[:n]

    # Gera o histograma. O 'density=True' garante que a área total seja 1.
    plt.figure(figsize=(8, 5))
    plt.hist(sub_amostras, bins=30, density=True, alpha=0.7)
    plt.title(f'Histograma com {n} pontos')
    plt.xlabel('Valor')
    plt.ylabel('Densidade')
    plt.grid(True)
    plt.show()

# --- Seção b: Calcular e Analisar Média e Variância ---
print("\n--- Análise de Média e Variância ---")

medias = []
variancias = []
num_pontos_list = []

for n in amostras_a_analisar:
    # Seleciona o subconjunto de amostras
    sub_amostras = amostras_normais[:n]

    # Calcula a média e a variância
    media_amostral = np.mean(sub_amostras)
    variancia_amostral = np.var(sub_amostras)

    # Armazena os resultados para o gráfico posterior
    num_pontos_list.append(n)
    medias.append(media_amostral)
    variancias.append(variancia_amostral)

    print(f"Para n = {n} pontos:")
    print(f"  Média amostral = {media_amostral:.4f}")
    print(f"  Variância amostral = {variancia_amostral:.4f}")

# Plotar os gráficos da média e da variância em função do número de pontos
plt.figure(figsize=(10, 6))
plt.plot(num_pontos_list, medias, 'o-', label='Média Amostral')
plt.axhline(y=0, color='r', linestyle='--', label='Média Teórica (μ=0)')
plt.title('Média Amostral em função do Número de Pontos')
plt.xlabel('Número de Pontos')
plt.ylabel('Média')
plt.xscale('log') # Escala logarítmica para visualizar melhor a convergência
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(num_pontos_list, variancias, 'o-', label='Variância Amostral')
plt.axhline(y=1, color='r', linestyle='--', label='Variância Teórica (σ²=1)')
plt.title('Variância Amostral em função do Número de Pontos')
plt.xlabel('Número de Pontos')
plt.ylabel('Variância')
plt.xscale('log')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da simulação
num_amostras = 5000
media_teorica = 0
desvio_padrao_teorico = 1

# Parâmetros da questão para a, b e C
a_param = 1
b_params = [-10, -1, -0.1, 0, 0.1, 1, 10]
C_params = [0, 5]

# Gerar as amostras de X e Y (fixas para todas as combinações)
X = np.random.normal(loc=media_teorica, scale=desvio_padrao_teorico, size=num_amostras)
Y = np.random.normal(loc=media_teorica, scale=desvio_padrao_teorico, size=num_amostras)

print("--- Análise da Covariância e Estatísticas de Z ---")

for C in C_params:
    for b in b_params:
        # Calcular Z com os parâmetros atuais
        Z = a_param * X + b * Y + C

        # --- Cálculo das Estatísticas ---

        # Simulado
        cov_XY_simulado = np.cov(X, Y)[0, 1]
        cov_XZ_simulado = np.cov(X, Z)[0, 1]
        media_Z_simulada = np.mean(Z)
        variancia_Z_simulada = np.var(Z)

        # Teórico
        cov_XY_teorico = 0
        cov_XZ_teorico = a_param
        media_Z_teorica = C
        variancia_Z_teorica = a_param**2 + b**2

        print(f"\nParâmetros: a={a_param}, b={b}, C={C}")
        print(f"  Cov(X,Y): Simulada = {cov_XY_simulado:.4f}, Teórica = {cov_XY_teorico}")
        print(f"  Cov(X,Z): Simulada = {cov_XZ_simulado:.4f}, Teórica = {cov_XZ_teorico}")
        print(f"  E[Z]: Simulada = {media_Z_simulada:.4f}, Teórica = {media_Z_teorica}")
        print(f"  Var(Z): Simulada = {variancia_Z_simulada:.4f}, Teórica = {variancia_Z_teorica}")

        # --- Gráfico de Dispersão (Scatter Plot) de (X, Z) ---
        plt.figure(figsize=(8, 6))
        plt.scatter(X, Z, alpha=0.5, s=10) # 's=10' para tamanho do ponto
        plt.title(f'Gráfico de Dispersão de (X, Z) para a={a_param}, b={b}, C={C}')
        plt.xlabel('X')
        plt.ylabel('Z')
        plt.grid(True)
        plt.show()
```
