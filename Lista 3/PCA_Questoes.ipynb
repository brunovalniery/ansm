```
# Quest達o 1 - PCA Digits
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

digits = load_digits()
X = digits.data
y = digits.target

sc = StandardScaler()
X_std = sc.fit_transform(X)

pca2 = PCA(n_components=2)
X_pca2 = pca2.fit_transform(X_std)

plt.scatter(X_pca2[:,0], X_pca2[:,1], c=y, cmap='tab10', s=10)
plt.show()
```

```
# Quest達o 1 - PCA Cancer
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

cancer = load_breast_cancer()
X = cancer.data
y = cancer.target

sc = StandardScaler()
X_std = sc.fit_transform(X)

pca2 = PCA(n_components=2)
X_pca2 = pca2.fit_transform(X_std)

plt.scatter(X_pca2[:,0], X_pca2[:,1], c=y, cmap='viridis', s=20)
plt.show()
```

```
# Quest達o 2 - PCA Curva S
from sklearn.datasets import make_s_curve
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X, color = make_s_curve(2000, random_state=0)
sc = StandardScaler()
X_std = sc.fit_transform(X)

pca2 = PCA(n_components=2)
X_pca2 = pca2.fit_transform(X_std)

plt.scatter(X_pca2[:,0], X_pca2[:,1], c=color, cmap='jet', s=10)
plt.show()
```
